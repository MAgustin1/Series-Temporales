```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.ticker import IndexLocator
import seaborn as sns
```

```python
df1 = pd.read_csv('desempleo de usa.csv')
```

```python
fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(df1["DATE"],df1["UNRATE"],label="Desempleo")
plt.title("Medici√≥n del desempleo en USA 1948 - 2023")
plt.xlabel("Fecha")
plt.ylabel("Millones")
plt.yticks(range(2,16,2))


locator = IndexLocator(base=40, offset= 20)
ax.xaxis.set_major_locator(locator)

ax.tick_params(axis='x', labelrotation=45)

plt.legend()
plt.show()
```
